```python
import numpy as np
import matplotlib.pyplot as plt

def clt(sample_size):
    data_mean = np.empty(10000)

    for i in range(10000):
        data_generate = np.random.exponential(scale=1, size=sample_size)
        data_mean[i] = np.mean(data_generate)

    plt.figure(figsize=(12, 6))

    # Plot population histogram
    plt.subplot(1, 2, 1)
    plt.hist(data_generate, bins=30, edgecolor='black')
    plt.title('Population')

    # Plot sample average histogram
    plt.subplot(1, 2, 2)
    plt.hist(data_mean, bins=30, edgecolor='black')
    plt.title('Sample Average')

    plt.show()

    print('The theoretical mean and variance of the population is 1')
    print('The estimated mean of the sample average is:', np.mean(data_mean))
    print('The estimated variance of the sample is:', np.var(data_mean))

# Test the function with different sample sizes
#clt(5)
#clt(10)
#clt(30)
#clt(50)
#clt(100)
clt(1000)
```
